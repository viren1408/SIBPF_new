```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
# Read the pulsar candidates
candidates_day_1 = pd.read_csv('Pulsar_candidates_G033.0-5.0.csv')


#Step 1 : Extract the x_pos and y_pos for the candidates in the image : 
#x_pos_day_1 = candidates_day_1[' Xposn']
#y_pos_day_1 = candidates_day_1[' Yposn']
#Peak_flux_day_1 = candidates_day_1[' Peak_flux']
# Assuming 'source_mask' represents the approximate source size in degrees
#source_mask = candidates_day_1[' Maj'] + candidates_day_1[' Min']

# Load FITS file for day 2
day_2_fits_file = 'G033.0-5.0_mod2.FITS'
day_2_hdul = fits.open(day_2_fits_file)
day_2_data = day_2_hdul[0].data
day_2_header = day_2_hdul[0].header
day_2_hdul.close()

cdelt1 = day_2_header['CDELT1']
cdelt2 = day_2_header['CDELT2']
pixel_scale_degrees_per_pixel = np.sqrt(cdelt1**2 + cdelt2**2)




# Extract pixel values at candidate positions within the defined source size
candidate_pixel_values = []

for i in range(len(candidates_day_1)): 
	x_pos_day_1 = candidates_day_1[' Xposn'][i]
	y_pos_day_1 = candidates_day_1[' Yposn'][i]
	Peak_flux_day_1 = candidates_day_1[' Peak_flux'][i]
	source_mask = (candidates_day_1[' Maj'][i] + candidates_day_1[' Min'][i])/2 #Source size in degrees
	source_mask_pixels = source_mask / pixel_scale_degrees_per_pixel # Convert source size from degrees to pixels
	#Source extraction 
	# Print out debugging information
	print(f"Candidate {i+1}: X={x_pos_day_1}, Y={y_pos_day_1}, Source Mask (degrees)={source_mask}, Source Mask (pixels)={source_mask_pixels}")
	
	# Calculate source boundaries
	x_start = max(int(x_pos_day_1 - source_mask_pixels / 2), 0)
	x_end = min(int(x_pos_day_1 + source_mask_pixels / 2), day_2_data.shape[3])
	y_start = max(int(y_pos_day_1 - source_mask_pixels / 2), 0)
	y_end = min(int(y_pos_day_1 + source_mask_pixels / 2), day_2_data.shape[2])
    
    	# Extract source values
	source_values = day_2_data[0, 0, y_start:y_end, x_start:x_end]
    
	candidate_pixel_values.append(source_values)

	# Print pixel values for the current candidate
	print(f'Candidate {i+1} Pixel Values:')
	# Choose a candidate index for visualization (e.g., candidate 7)

	# Get the flux values for the chosen candidate
	flux_values = candidate_pixel_values[i]

	# Create a heatmap (2D array plot) of the flux values
	plt.figure(figsize=(8, 6))
	plt.imshow(flux_values, cmap='viridis', origin='lower', aspect='auto')
	plt.colorbar(label='Flux')
	# Get the x and y coordinates of the region
	x_coords = np.arange(x_start, x_end)
	y_coords = np.arange(y_start, y_end)

# Display actual pixel coordinates on the x and y axes
	plt.xticks(ticks=np.arange(len(x_coords)), labels=x_coords)
	plt.yticks(ticks=np.arange(len(y_coords)), labels=y_coords)

	plt.xlabel('X Coordinate (pixels)')
	plt.ylabel('Y Coordinate (pixels)')
	plt.title(f'Flux Distribution for Candidate {i}')
	plt.show()

	print(source_values)




#print(f'All Pixel Values: {candidate_pixel_values}')


```
